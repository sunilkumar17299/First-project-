```python
ratings=pd.read_csv("/content/Ratings.csv")
ratings.head()
```

```python
x=ratings["User-ID"].value_counts()>=200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings["User-ID"].isin(y)]
```

```python
ratings.shape
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books,on="ISBN")
```

```python
ratings_with_books.head()
ratings_with_books.shape
```

```python
ratings_with_books.head()
```

```python
number_rating=ratings_with_books.groupby("Book-Title")["Book-Rating"].count().reset_index()
number_rating.head()
```

```python
final_rating=ratings_with_books.merge(number_rating,on="Book-Title")
final_rating.head()
```

```python
final_rating=final_rating.rename(columns={"Book-Rating_x":"Book-Rating","Book-Rating_y":"number_of_ratings"})
final_rating.head()
```

```python
final_rating=final_rating[final_rating["number_of_ratings"]>=50]
final_rating.head()
```

```python
print(final_rating.columns)
final_rating.head()
```

```python
print(final_rating.columns)
```

```python
book_pivot=final_rating.pivot_table(columns="User-ID",index="Book-Title",values="Book-Rating")
book_pivot.head()
```

```python
book_pivot.fillna(0,inplace=True)
book_pivot.head()
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_sparse)
```

```python
book_pivot.index==237
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
  for i in range(len(suggestions)):
    if i==0:
      print("The suggestions for",book_name,"are:")
    if not i:
      print(book_pivot.index[suggestions[i]])


```

```python
recommend_book("Animal Farm")
```
